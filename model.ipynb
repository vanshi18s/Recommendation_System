```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
df=pd.read_csv('preprocessed_data.csv')
```

```python
cols_to_combine = ['screen_time_min', 'launches', 'interactions']
scaler = MinMaxScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df[cols_to_combine]), columns=cols_to_combine)
df['combined_rating'] = df_scaled.mean(axis=1)
```

```python
user_item_matrix = df.pivot_table(
    index='user_id',
    columns='app_name',
    values='combined_rating',
    fill_value=0 )
print(user_item_matrix)
```

```python
item_similarity = cosine_similarity(user_item_matrix.T)  
item_similarity_df = pd.DataFrame(item_similarity, index=user_item_matrix.columns, columns=user_item_matrix.columns)
```

```python
item_similarity_df.head(25)
```

```python
import pandas as pd

def recommend_apps(
    user_id, 
    user_item_matrix, 
    item_similarity_df, 
    N=5
):
    if user_id not in user_item_matrix.index:
        print(f"User ID {user_id} not found in data.")
        return []
    user_ratings = user_item_matrix.loc[user_id]
    used_apps = user_ratings[user_ratings > 0].index
    unused_apps = user_ratings[user_ratings == 0].index
    
    scores = {}
    for app in unused_apps:
        sim_vec = item_similarity_df.loc[app, used_apps]
        rating_vec = user_ratings[used_apps]
        if sim_vec.sum() > 0:
            score = (sim_vec * rating_vec).sum() / sim_vec.sum()
        else:
            score = 0
        scores[app] = score
    recommendations = sorted(scores.items(), key=lambda x: x[1], reverse=True)[:N]
    return recommendations
```

```python
#Testing Predictions:

u_id=int(input("Please enter the user_id of the user you want to make recommendations for: "))
top_recommendations = recommend_apps(u_id, user_item_matrix=user_item_matrix, item_similarity_df=item_similarity_df, N=5)
print(top_recommendations)
```

```python
# Model Evaluation
from sklearn.metrics import mean_squared_error
```

```python
def evaluate_model(user_item_matrix, item_similarity_df, sample_users=10):
    actual_ratings = []
    predicted_ratings = []

    users = user_item_matrix.index.tolist()[:sample_users] 
    for user_id in users:
        user_ratings = user_item_matrix.loc[user_id]
        used_apps = user_ratings[user_ratings > 0].index.tolist()
        if len(used_apps) < 2:
            continue
        test_app = used_apps[-1]
        true_rating = user_ratings[test_app]
        user_item_matrix.loc[user_id, test_app] = 0
        recs = recommend_apps(user_id, user_item_matrix, item_similarity_df, N=len(user_item_matrix.columns))
        predicted = dict(recs).get(test_app, None)
        user_item_matrix.loc[user_id, test_app] = true_rating
        if predicted is not None:
            actual_ratings.append(true_rating)
            predicted_ratings.append(predicted)

    if actual_ratings:
        rmse = np.sqrt(mean_squared_error(actual_ratings, predicted_ratings))
        print(f"RMSE: {rmse:.2f}")
    else:
        print("Not enough data to evaluate.")
```

```python
evaluate_model(user_item_matrix, item_similarity_df, sample_users=10)
```
